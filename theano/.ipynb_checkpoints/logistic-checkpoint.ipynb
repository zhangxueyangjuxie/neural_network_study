```python
import numpy as np
import theano
import theano.tensor as T
```

```python
R = np.random
N = 10     # 我们为了测试，自己生成10个样本，每个样本是3维的向量，然后用于训练  
feats = 3  
D = (R.randn(N, feats).astype(np.float32), R.randint(size=N, low=0, high=2).astype(np.float32))  
```

```python
x = T.matrix('x')
y = T.vector('y')
```

```python
w = theano.shared(R.randn(feats), name = 'w')
b = theano.shared(0., name='b')
```

```python
p_1 = 1 / (1 + T.exp(-T.dot(x,w)  - b))
xent = - y * T.log(p_1) - (1-y) * T.log(1-p_1)
cost = xent.mean() + 0.01 * (w**2).sum()
gw, gb = T.grad(cost, [w, b])
```

```python
prediction = p_1 > 0.5
train = theano.function(inputs = [x, y], outputs = [prediction, xent], updates =
                       ((w, w - 0.1 * gw), (b, b - 0.1 * gb)))
predict = theano.function(inputs = [x], outputs = prediction)
```

```python
training_steps = 1000
for i in range(training_steps):
    pred, err = train(D[0],D[1])
    print err.mean()
```
